```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_excel('testedatasetv2b.xlsx')

df['V'] = df['V'].fillna(0)
df['E'] = df['E'].fillna(0)
df['D'] = df['D'].fillna(0)
df['GOL'] = df['GOL'].fillna(0)
df['ASS'] = df['ASS'].fillna(0)
df['STG'] = df['STG'].fillna(0)
df['GC'] = df['GC'].fillna(0)
df['AMA'] = df['AMA'].fillna(0)
df['AZUL'] = df['AZUL'].fillna(0)
df['VER'] = df['VER'].fillna(0)
df['PP'] = df['PP'].fillna(0)
df['GS'] = df['GS'].fillna(0)
df['DD'] = df['DD'].fillna(0)
df['DP'] = df['DP'].fillna(0)
df['FALTA'] = df['FALTA'].fillna(0)
```

```python
df.head(10)
```

```python
df_fev = df.loc[(df['RODADA'] >= 4) & (df['RODADA']<=6)]  
df_fev = df_fev.drop(["PARTIDA", "RODADA", "PARTIDA", "POSIÇÃO", "TIME"], axis=1)
df_fev = df_fev.groupby('PLAYER').sum()
df_fev = df_fev.reset_index()
df_fev = df_fev.sort_values('PTS', ascending=False)
df_fev
```

```python
df_round_esp= df.loc[df['RODADA']==7]

df_round_esp = df_round_esp.drop(["PARTIDA", "RODADA", "PARTIDA", "POSIÇÃO", "TIME"], axis=1)
df_round_esp = df_round_esp.groupby('PLAYER').sum().reset_index()
df_round_esp = df_round_esp.sort_values('PTS', ascending=False)
df_round_esp
```
